```python
# No statistically significant difference in sex distribution - initial

import numpy as np
from scipy.stats import chi2_contingency
import pandas as pd


data = pd.DataFrame({
    "Female": [15, 19],
    "Male": [33, 18]
}, index=["MND", "Control"])


# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(data, correction=False) # yates uncorrected as expected count > 5 


# Print the results
print(f"Chi-square statistic: {chi2:.2f}")
print(f"p-value: {p:.3f}")
print("Expected counts:\n", expected)


# Interpretation
if p < 0.05:
    print("There is a statistically significant difference in sex distribution between the groups.")
else:
    print("There is no statistically significant difference in sex distribution between the groups.")

print(data)
```
